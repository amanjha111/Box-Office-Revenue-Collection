```python
df_new=pd.read_csv("/content/movie_prediction.csv")
df_new
```

```python
Categorical_columns=df_new.select_dtypes(include=['object']).columns.tolist()
Categorical_columns
```

```python
from sklearn.preprocessing import OneHotEncoder
enconder=OneHotEncoder(sparse_output=False, handle_unknown='ignore')
one_hot_encoding=enconder.fit_transform(df_new[Categorical_columns].astype(str))
one_hot_encoding
```

```python
newdf=pd.DataFrame(one_hot_encoding,columns=enconder.get_feature_names_out())
newdf
```

```python
df_encoded=pd.concat([df_new,newdf],axis=1)
df_encoded
```

```python
cat_columns=df_encoded.select_dtypes(include=['object']).columns.tolist()
print(cat_columns)
```

```python
df_encoded.isnull().sum().sum()
```

```python
ds=df_encoded.drop(columns=cat_columns)
ds
```

```python
x=ds.drop(columns=['index'])
y=ds['index']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_R= KNeighborsRegressor(n_neighbors=5)
knn_R.fit(x_train, y_train)
```

```python
knn_R.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_R1= KNeighborsRegressor(n_neighbors=3)
knn_R1.fit(x_train, y_train)
```

```python
knn_R1.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_R2= KNeighborsRegressor(n_neighbors=7)
knn_R2.fit(x_train, y_train)
```

```python
knn_R2.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_C = KNeighborsClassifier(n_neighbors=7)
knn_C.fit(x_train, y_train)
```

```python
knn_C.score(x_test, y_test)
```
